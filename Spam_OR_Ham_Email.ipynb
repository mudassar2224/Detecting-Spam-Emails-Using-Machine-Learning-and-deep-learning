```python
def remove_stopwords(text):
  stop_words=stopwords.words("english")
  imp_words=[]
  for word in  str(text).split():
    word=word.lower()
    if word in stop_words:
      imp_words.append(word)
      output=" ".join(imp_words)
    return output
  balanced_data["text"]=balanced_data["text"].apply(lambda text: remove_stopwords(text))
  balanced_data.head()
```

# ***Step No.06*** `Tokenization And Padding`

**What’s Going On?***

After tokenization, your sentences are now lists of numbers (word IDs).

But the problem is:
 Every sentence has a different number of words!

Neural networks require inputs of the same length, just like a fixed-size matrix.
That’s where pad_sequences() comes in.

```python
X_train,X_test, y_train , y_test=train_test_split(
    balanced_data["text"], balanced_data["label"], test_size=0.2, random_state=42

)
tokenizer=Tokenizer()
tokenizer.fit_on_texts(X_train)
train_sequences=tokenizer.texts_to_sequences(X_train)
test_sequences=tokenizer.texts_to_sequences(X_test)
max_len=100
train_sequences=pad_sequences(
    train_sequences, maxlen=max_len, padding="post", truncating="post"
)
test_sequences=pad_sequences(
  test_sequences, maxlen=max_len, padding="post", truncating="post"
)
y_train=(y_train=="spam").astype(int)
y_test=(y_test=="spam").astype(int)
```

# ***🧰 pad_sequences() — What It Does***

`It makes all sequences the same length by:`

`Adding zeros (padding) if a sentence is too short`

`Truncating (cutting) if a sentence is too long`

# ***Define Model***

# ***🔍 Purpose:***

The Embedding Layer converts word IDs (integers) into dense numeric vectors — small sets of numbers that capture meaning of words

| Word    | ID | Embedding (32 numbers example) |
| ------- | -- | ------------------------------ |
| “free”  | 1  | `[0.12, -0.45, 0.09, ...]`     |
| “offer” | 2  | `[0.33, 0.02, -0.18, ...]`     |


| Parameter      | Meaning                                                                         |
| -------------- | ------------------------------------------------------------------------------- |
| `input_dim`    | Total number of words (vocabulary size) = number of unique words from Tokenizer |
| `output_dim`   | Dimension of each word vector (here 32 means every word → 32 numbers)           |
| `input_length` | Length of each input sequence (you set `max_len = 100`)                         |


***LSTM***
LSTM = Long Short-Term Memory, a type of Recurrent Neural Network (RNN).

It reads the sequence word by word and learns the context / meaning of the whole sentence.
For example:

“You won a free iPhone” → LSTM learns that “won” + “free” = spam context.

***Dense Layer***
Fully connected hidden layer that learns nonlinear patterns.

relu (Rectified Linear Unit) means:

Keep positive values as-is

Set negative values to 0
This helps the network learn efficiently.

It adds another transformation step before the final decision.

# ***Sigmoid***

Purpose:

Outputs a single value between 0 and 1:

Closer to 1 → more likely “spam”

Closer to 0 → more likely “ham”

sigmoid activation compresses outputs into [0, 1] range (useful for binary classification)

```python
model=tf.keras.models.Sequential([
    tf.keras.layers.Embedding(input_dim=len(tokenizer.word_index)+1, output_dim=32, input_length=max_len
     ),
    tf.keras.layers.LSTM(16, unroll=True),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(32, activation="relu"),
    tf.keras.layers.Dense(1, activation="sigmoid")
])
model.compile(
    loss=tf.keras.losses.BinaryCrossentropy(),
    optimizer="adam",
    metrics=["accuracy"]

)

model.summary()
```

# ***Loss And BinaryCross Enotropy***

`🔹 What is “loss”?`

The loss function measures how wrong your model’s predictions are.

The lower the loss, the better your model is learning.

`🔹 Why “BinaryCrossentropy”?`

Because this is a binary classification problem — spam (1) or ham (0).

It’s the standard loss function for problems with two possible outputs.

#***Adam***

`Why “Adam”?`

Adam = Adaptive Moment Estimation.
It’s one of the best general-purpose optimizers — it:

Automatically adjusts learning rates

Works well for text, images, and tabular data

Converges faster than basic optimizers like SGD

So, 'adam' is an excellent choice here

` metrics=['accuracy']`

This tells TensorFlow to show accuracy while training —
how many predictions are correct.

# |***`Train Model`***

| Callback            | Purpose                           | Example Effect            |
| ------------------- | --------------------------------- | ------------------------- |
| `EarlyStopping`     | Stop training when no improvement | Prevents overfitting      |
| `ReduceLROnPlateau` | Lower learning rate if stuck      | Helps reach better minima |


💡 Analogy:

You’re climbing a mountain ⛰️ (minimizing loss).

If you stop progressing, ReduceLROnPlateau says: “Slow down and take smaller steps.”

If you still can’t improve, EarlyStopping says: “Okay, stop climbing — you’ve reached your best point.”

```python
es=EarlyStopping(patience=3, monitor="val_accuracy", restore_best_weights=True)
lr=ReduceLROnPlateau(patience=2, monitor="val_loss", factor=0.5, verbose=0)
history=model.fit(train_sequences,y_train,validation_data=(test_sequences, y_test) ,epochs=20, batch_size=32,callbacks=[lr , es])
```

```python
test_loss, test_accuracy = model.evaluate(test_sequences, y_test)
print('Test Loss :',test_loss)
print('Test Accuracy :',test_accuracy)
```

```python
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```
